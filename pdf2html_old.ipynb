```python
#first download tesseract to use
#acrobat > export pdf > image > tiff > check box for "export all images" > settings (gear icon) > under convert: select grayscale, 300 dpi

from PIL import Image
import pytesseract
import glob
import warnings
warnings.filterwarnings('ignore')

tifs = glob.glob('glossary/*.tiff')
tifs = [i for i in tifs if '0002' in i] #if duplicates
tifs = sorted(tifs)
```

```python
## test 1 page :
img = Image.open(tifs[2])
img.load()
text = pytesseract.image_to_string(img)
print(text)
```

```python
language = 'eng'
full = ''
for x in tifs:
    img = Image.open(x)
    img.load()
    text = pytesseract.image_to_string(img)
    full = full + text
    print('Page', tifs.index(x), 'of', len(tifs), 'DONE')
```

```python
with open('opencity_full.html', 'x', encoding='utf-16') as file:
    file.write(full)
```
