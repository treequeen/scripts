```python
import requests
import json
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.0f' % x)
```

```python
url = 'https://bikeindex.org:443/api/v3/search?page=1&per_page=100&location=Boston%2C%20MA&distance=5&stolenness=proximity'
r = requests.get(url)
result = r.json()
bikes = pd.DataFrame(result['bikes'])
bikes
```

```python
for i in range(2,20):
    pgno = i
    url = f'https://bikeindex.org:443/api/v3/search?page={pgno}&per_page=100&location=Boston%2C%20MA&distance=5&stolenness=proximity'
    r = requests.get(url)
    result = r.json()
    df = pd.DataFrame(result['bikes'])
    bikes = pd.concat([bikes, df],ignore_index=True)
```

```python
bikes
```

```python
bikes.to_csv("geodata/bikeindex.csv")
```
